```python
df=df.drop(['Unnamed: 0','Age','subject_id','video_id','Education Level','Title','Fields of Interest'],axis=1)
```

```python
df['subject_understood'].replace(['Not Understood', 'Understood'],
                        [0, 1], inplace=True)
```

```python
import numpy as np
from scipy import stats
df_zscore = df[(np.abs(stats.zscore(df)) < 3).all(axis=1)]
```

```python
df_zscore.shape
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 8))
dp = sns.boxplot(x= 'POW.T7.Theta' ,data = df, ax = ax[0])
ax[0].set_title('without zscore')
dp = sns.boxplot(x= 'POW.T7.Theta' ,data = df_zscore, ax = ax[1])
ax[ 1].set_title('With zscore') 
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 8))
dp = sns.boxplot(x= np.log2(df['POW.T7.Theta']), ax = ax[0])
ax[0].set_title('without zscore with log tranformation')
dp = sns.boxplot(x= np.log2(df_zscore['POW.AF4.Alpha']), ax = ax[1])
ax[ 1].set_title('With zscore with log tranformation') 
plt.show()
```

```python
X= df_zscore.drop('subject_understood', axis=1)
Y= df_zscore['subject_understood']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=200, random_state=0)
rfc.fit(X_train, Y_train)
y_pred = rfc.predict(X_test)
```

```python
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Y_test, y_pred)))
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(Y_test,y_pred))
print(classification_report(Y_test,y_pred))
print(accuracy_score(Y_test, y_pred))
```

```python
# Making the Confusion Matrix
cm = confusion_matrix(Y_test,y_pred)
#print(cm)
# create heatmap
sns.heatmap(pd.DataFrame(cm), annot=True, cmap="flare" ,fmt='g')

plt.tight_layout()
plt.title('Confusion matrix for KNN', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
print("Model Accuracy for Random Forest:",accuracy_score(Y_test, y_pred) * 100)
```
