```python
df=df.drop(['Unnamed: 0','Age','subject_id','video_id','Education Level','Title','Fields of Interest'],axis=1)
```

```python
df['subject_understood'].replace(['Not Understood', 'Understood'],
                        [0, 1], inplace=True)
```

```python
from scipy import stats
df_zscore = df[(np.abs(stats.zscore(df)) < 3).all(axis=1)]
```

```python
import pandas as pd

def remove_outliers(df,columns,n_std):
    for col in columns:
        print('Working on column: {}'.format(col))
        
        mean = df[col].mean()
        sd = df[col].std()
        
        df = df[(df[col] <= mean+(n_std*sd))]
        
    return df
```

```python
columns = df.columns
n_std = 3
df_no_outliers = remove_outliers(df_zscore, columns, n_std)
```

```python
sns.boxplot(x = 'POW.AF4.BetaH', data=df_no_outliers)
```

```python
sns.kdeplot(x = 'POW.AF4.Alpha', data = df_no_outliers)
```

```python
sns.boxplot(x = 'POW.AF4.Alpha', data= df_no_outliers)
```

```python
X = df_no_outliers.drop('subject_understood', axis=1)

Y = df_no_outliers['subject_understood']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn import datasets, linear_model, metrics
reg = linear_model.LogisticRegression()
reg.fit(X_train, Y_train)
y_pred = reg.predict(X_test)
```

```python
print("Logistic Regression model accuracy(in %):", 
metrics.accuracy_score(Y_test, y_pred)*100)
```

```python
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Y_test, y_pred)))
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(Y_test,y_pred))
print(classification_report(Y_test,y_pred))
print(accuracy_score(Y_test,y_pred))
```
